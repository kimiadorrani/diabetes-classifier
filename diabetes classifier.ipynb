```python
import numpy as np
import pandas as pd
```

here we import our dataset

```python
diabetes = pd.read_csv('diabetes.csv')
diabetes
```

with following command we are able to see null values in our data set so we can clean it.

```python
diabetes.isna().sum()
```

Here we want to handle null values .


In columns 'BMI' , 'Mental Health' , 'Physical Health' we replace null values with mean of their values.(Replacing With Mean)


In columns with categorical values we replace them with 'missing' label.(Impute the Value “missing”, which treats it as a Separate Category)


In other columns which only have 0/1 or male/female, we replace them with 0/male. which is one of their values.(Impute the Most Frequent Value)

```python
diabetes["BMI"].fillna(diabetes["BMI"].mean(), inplace=True)
diabetes["HeartDiseaseorAttack"].fillna(0, inplace=True)
diabetes["Physical Activity"].fillna(1, inplace=True)
diabetes["Fruits"].fillna(1, inplace=True)
diabetes["Veggies"].fillna(1, inplace=True)
diabetes["Heavy Alcohol Consumption"].fillna(0, inplace=True)
diabetes["Any Health Care"].fillna(1, inplace=True)
diabetes["No Doctor because of Cost"].fillna(0, inplace=True)
diabetes["General Health"].fillna("missing", inplace=True)
diabetes["Mental Health"].fillna(diabetes["Mental Health"].mean(), inplace=True)
diabetes["Physical Health"].fillna(diabetes["Physical Health"].mean(), inplace=True)
diabetes["Difficulty Walking"].fillna(0, inplace=True)
diabetes["Sex"].fillna("female", inplace=True)
diabetes["Age"].fillna(diabetes["Age"].mean(), inplace=True)
diabetes["Education"].fillna("missing", inplace=True)
diabetes["Income"].fillna("missing", inplace=True)

```

Here we find out that there is an unknown value in Income which in nonesense.

```python
diabetes['Income'].value_counts()
```

Here we just delete Unknown from data frame

```python
diabetes = diabetes[diabetes['Income'] != 'Unknown']
diabetes['Income'].value_counts()
```

```python
diabetes.isna().sum()
```

```python
diabetes.columns
```

Here we want to rename our column names which they have space in them and replace it with underline.

```python
diabetes.columns = diabetes.columns.str.replace(' ', '_')
diabetes
```

Here we drop Unnamed:_0 column which is nonesense and can cause noise

```python
diabetes.drop("Unnamed:_0",axis=1,inplace=True)
diabetes
```

```python
diabetes.info()
```

Here we will normalize our data in columns that have values greater than 5 and map them to the range of 0-1.

```python
pd.options.mode.chained_assignment = None  # default='warn'

diabetes["Mental_Health"] = diabetes["Mental_Health"] / diabetes["Mental_Health"].max()
diabetes["Physical_Health"] = diabetes["Physical_Health"] / diabetes["Physical_Health"].max()
diabetes["BMI"] = diabetes["BMI"] / diabetes["BMI"].max()
diabetes["Age"] = diabetes["Age"] / diabetes["Age"].max()
diabetes
    
```

Here we want to handle categorical features using one hot encoder as mentioned.

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
X = diabetes.drop("Diabetes_binary", axis=1)
y = diabetes["Diabetes_binary"]
categorical_features = ["General_Health", "Sex", "Education", "Income"]
one_hot = OneHotEncoder()
transformer = ColumnTransformer([("one_hot", 
                                 one_hot, 
                                 categorical_features)],
                                 remainder="passthrough")
transformed_X = transformer.fit_transform(X)
transformed_X
transformed_X[0]
```

Here we will split our data to two sets of Train and Test set with size of 0.2 as it was mentioned in class.


it means that we use 80 percent of our data to train and 20 percent for testing.

```python
from sklearn.model_selection import train_test_split

np.random.seed(42)
X_train, X_test, y_train, y_test = train_test_split(transformed_X, y, test_size=0.2)

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

Here we wanted to define a XGboost classifier and it parameters was mentioned in project template.

```python
from xgboost import XGBClassifier

# fit model no training data
model = XGBClassifier(learning_rate=0.1, max_depth=4, n_estimators=200, subsample=0.5, colsample_bytree=1, random_seed=123, eval_metric='auc',verbosity=1)
model.fit(X_train, y_train,eval_set=[(X_test, y_test)],early_stopping_rounds=10)
```

Here we just want to score our classifier with our test data:

```python
model.score(X_test, y_test)
```

Here we import classification_report in order to have precision and recall.


y_pred is what we get as an output for X_test and then we get classification_report.

```python
from sklearn.metrics import classification_report
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

here we want to see confusion matrix. we just used heatmap in order to see this confiusion matrix in graphical form.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
table=sns.heatmap(confusion_matrix(y_test,y_pred),annot=True, fmt='g',cmap='Blues')
table.xaxis.set_ticklabels(['False','True'])
table.yaxis.set_ticklabels(['False','True'])
```

here we test model on different combination of parameters and as an output we will get the best model and estimator

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import roc_auc_score

def my_roc_auc_score(model, X, y): return roc_auc_score(y, model.predict_proba(X)[:,1])

parameters = {
    'learning_rate' : [0.02,0.05,0.1,0.3],
    'max_depth' : [2,3,4],
    'n_estimators' : [100,200,300],
    'colsample_bytree' : [0.8,1]
}

estimator = XGBClassifier(
    objective= 'binary:logistic',
    nthread=4,
    seed=42
)
grid_search = GridSearchCV(
    estimator=estimator,
    param_grid=parameters,
    scoring = my_roc_auc_score,
    cv = 3
    )

grid_search.fit(X_train, y_train) 
print("Best score: %f using %s" % (grid_search.best_score_, grid_search.best_params_))

```

Best estimator score

```python
grid_search.best_estimator_.score(X_test, y_test)
```

Here we also see precison and recall for the mentioned model.

```python
grid_predict = grid_search.predict(X_test) 
   
# print classification report 
print(classification_report(y_test, grid_predict))
```

and here we see confusion matrix.

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
table=sns.heatmap(confusion_matrix(y_test,grid_predict),annot=True, fmt='g',cmap='Blues')
table.xaxis.set_ticklabels(['False','True'])
table.yaxis.set_ticklabels(['False','True'])
```
